1)  What is Python, and why is it popular?

-->Python is a versatile, high-level, interpreted programming language known for its readability and ease of use
and its popular because python have extensive libraries,simple syntax,and adaptability to various fields like web development, data science, and machine learning.

2) What is an interpreter in Python?

-->a program that reads and executes Python code line by line, without requiring it to be compiled into machine code beforehand.

3) What are pre-defined keywords in Python?

-->Keywords are predefined, reserved words used in Python programming that have special meanings to the compiler.

4) Can keywords be used as variable names?

-->No, keywords cannot be used as variable names.

5)  What is mutability in Python?

-->Mutability in Python refers to an object's ability to be modified after its creation. If an object is mutable, its state or contents can be changed without creating a new object.

6) Why are lists mutable, but tuples are immutable?

--> tuples use parentheses instead of square brackets, and the items in tuples cannot be modified, but the items in lists can be modified.

7) What is the difference between “==” and “is” operators in Python?

-->Use **==** to compare the content of objects (strings, lists, numbers).
Use **is** to check if two variables are referring to the same object (comparing with None).

8)  What are logical operators in Python?

-->Logical operators in Python are used to combine or modify boolean values. There are three primary logical operators
(not,and,or).

9) What is type casting in Python?

-->the process of converting one data type into another.

10) What is the difference between implicit and explicit type casting?

-->Implicit type casting happens automatically when converting data types, while explicit type casting requires a manual instruction from the programmer to convert a value from one type to another.

11)  What is the purpose of conditional statements in Python?

-->Conditional statements in Python are used to execute certain blocks of code based on specific conditions.

12) How does the elif statement work?

-->The code block under the elif statement will be executed only if its condition is true.

13) What is the difference between for and while loops?

-->for loops are used when you know how many times you want to iterate or when you need to process each element of a sequence.while loops repeatedly executed a block  of code until a condition is met.

14) Describe a scenario where a while loop is more suitable than a for loop.
